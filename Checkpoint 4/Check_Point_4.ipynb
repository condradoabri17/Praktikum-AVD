### **Outliers Values**

Pada tahap ini, dilakukan pengecekan dengan tujuan untuk mendeteksi nilai ekstrem yang berbeda jauh dari mayoritas data, karena outlier dapat memengaruhi:
* Nilai rata-rata (mean)
* Standar deviasi
* Hasil visualisasi

```python
results = []

cols = ['price', 'mileage', 'tax', 'mpg', 'engineSize']

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df.style.format({"Persentase Outliers": "{:.2f}%"}))
```

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

Tujuan dari tahap ini adalah untuk membandingkan rata-rata harga mobil berdasarkan model untuk melihat model mana yang memiliki harga lebih tinggi dibanding yang lain.

Visualisasi: Bar Chart

```python
price_by_model = df.groupby('model')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=price_by_model.index, y=price_by_model.values, palette='viridis')
plt.title('Rata-rata Harga Berdasarkan Model BMW')
plt.xlabel('Model')
plt.ylabel('Rata-rata Harga')
plt.xticks(rotation=45)
plt.show()
```

Berdasarkan grafik perbandingan rata-rata harga mobil bekas BMW berdasarkan model, terlihat adanya perbedaan harga yang cukup signifikan . Model X7 memiliki rata-rata harga tertinggi, diikuti oleh 8 Series, M5, dan i8, yang termasuk dalam kategori kendaraan premium. Hal ini menunjukkan bahwa model premium memiliki nilai jual bekas yang lebih tinggi dibanding model lainnya.

Sebaliknya, model seperti Z3, 1 Series, dan 2 Series berada pada kelompok dengan rata-rata harga terendah. Model-model ini umumnya termasuk dalam kategori mobil yang kelas menengah atau kendaraan yang lebih lama, sehingga harga jualnya relatif lebih murah.

Perbedaan rata-rata harga ini menunjukkan bahwa model kendaraan merupakan salah satu faktor utama yang memengaruhi harga jual mobil bekas BMW.

Bagi dealer, hasil ini dapat digunakan untuk menentukan strategi pengadaan dan penetapan harga.
