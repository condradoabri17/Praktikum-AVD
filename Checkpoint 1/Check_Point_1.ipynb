# **Business Understanding**

## **Business Objective**

Dataset ini berisi sekumpulan data tentang mobil bekas BMW. Data ini memiliki variabel utama yaitu Model, Tahun Keluaran Mobil, Jarak tempuh, Jenis Bahan Bakar, Jenis Transmisi dan model mobil. Tujuan dari menganalisis data ini adalah:
*   Mengetahui Faktor-Faktor apa saja yang akan menentukan harga sebuah mobil BMW bekas
*   Membantu Dealer untuk menentukan harga jual mobil bekas BMW agar masuk akal dan tetap laku dipasaran.
*   Mengidentifikasi tren di pasaran. Contohnya adalah "apakah mobil diesel lebih laku? atau apakah mobil matic lebih mahal?"

Kesimpulan:
Tujuan utama data ini untuk memudahkan dealer dalam memprediksi dan menentukan harga jual mobil BMW bekas yang masuk akal dan kompetitif, serta memprediksi dan menentukan tren pasar BMW sehingga pengadaan mobil bekas BMW bisa lebih sesusai dengan kebutuhan pasar.

## **Assess Situation**


Dataset ini menampung data yang cukup besar yaitu 10781 baris sehingga akan memperkuat prediksi harga dan faktor yang mempengaruhi harga tersebut, karena semakin banyak data yang dianalisis, maka akan semakin akurat. Selain itu berdasarkan analisis Mayoritas mobil diproduksi sekitar tahun 2017 dan Kapasitas Mesin rata-rata 2.0L. Selain itu tidak ditemukan nilai pada baris data yang hilang di salah satu kolom variabel, sehingga prediksi bisa semakin akurat dikarenakan lengkapnya data didalam dataset ini.

Untuk situasi bisnis, pasar mobil bekas biasanya dipengaruhi oleh:
*   Semakin baru tahun mobil maka akan semakin tingi harga jualnya.
*   Semakin tinggi jarak tempuh maka harga akan semakin turun
*   Model mobil akan mempengaruhi harga mobil. Misalnya model yang premium seperti 5 series kemungkinan harga jualnya akan mahal.
*   Mobil dengan transmisi otomatis (mobil matic) biasanya lebih mahal dibanding mobil dengan transmisi manual.
*   Mobil dengan bahan bakar diesel bisa lebih murah atau mahal, tergantung target dan pasarnya.
*  faktor-faktor atau variabel diluar dataset ini bisa jadi akan mempengaruhi harga jual mobil, misalnya kondisi fisik, lokasi, warna, dsb

Kesimpulannya adalah menganalisis dataset ini akan mendukung prediksi harga jual mobil bekas BMW yang lebih akurat karena faktor atau variabel yang dianalisis dalam dataset ini adalah faktor-faktor yang memang menjadi pertimbangan dalam situasi bisnis pasar mobil bekas.
